```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_gradient_art(width, height):
    # Generate a gradient using NumPy linspace
    x = np.linspace(0, 1, width)
    y = np.linspace(0, 1, height)
    gradient = np.outer(y, x)

    # Apply color to the gradient
    red_channel = gradient
    green_channel = gradient[::-1]  # Reverse the gradient vertically
    blue_channel = np.ones_like(gradient)  # Constant blue channel

    # Combine the color channels to create the image
    image = np.stack([red_channel, green_channel, blue_channel], axis=-1)

    plt.imshow(image)
    
create_gradient_art(300,100)
```
