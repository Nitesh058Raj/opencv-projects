```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('Resources/river.jpg')
img2 = cv2.imread('Resources/tree-middle-of-lack.jpg')

#BGR to RBG
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
def combine_images(image1, image2):
    # Combine the images side by side
    combined_image = cv2.hconcat([image1, image2])

    # Save the combined image
    plt.imshow(combined_image)
    
combine_images(img1, img2)
```

```python
img1 = cv2.imread('Resources/river.jpg')
img2 = cv2.imread('Resources/tree-middle-of-lack.jpg')

#BGR to RBG
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

def swap_cropped_parts(image1, image2):

    # Crop a part from image2
    crop2 = image2[1000:3000, 2000:4000]
    image1[1000:3000, 2000:4000] = crop2
    
    return image1
    
    
img1 = swap_cropped_parts(img1, img2)
plt.imshow(img1)
```
